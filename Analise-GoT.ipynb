```python
%matplotlib inline
import pandas as pd
import chardet
```

```python
def get_encoding(filename):
    with open(filename, 'rb') as f:
        result = chardet.detect(f.read())
        
    return result
```

```python
battles_file = pd.read_csv('./Battles.csv')
character_deaths_file = pd.read_csv('./Character Deaths.csv')
character_predictions_file = pd.read_csv('./Character Predictions.csv')
screentimes_file = pd.read_csv('./Screentimes.csv', encoding=get_encoding('./Screentimes.csv')['encoding'])
```

# Morte por Popularidade

```python
characters_file = character_predictions_file.set_index('name').join(character_deaths_file.set_index('Name'))
deaths_by_year = characters_file.groupby('Death Year').count()[['isPopular']].plot.bar(rot = 0, figsize = (14,12))
_ = deaths_by_year.set(xlabel = 'Ano da morte', ylabel = 'No. mortes por ano'), deaths_by_year.legend(['É Popular'])
```

# Maiores Mortes por episodios com eventos - Por Ano

```python
majors_by_year = battles_file.groupby('year').sum()[["major_death", "major_capture"]].plot.bar(rot = 0, figsize = (14, 12))
_ = majors_by_year.set(xlabel = 'Ano', ylabel = 'No. mortes/eventos em captulos', ylim = (0, 10)), majors_by_year.legend(['Morte figura importante', 'Episodio com evento'])
```

# Maiores Mortes por episodios com eventos - Por cada região

```python
majors_data = battles_file.groupby('region').sum()[["major_death", "major_capture"]]
majors_by_region = pd.concat([majors_data, battles_file.region.value_counts().to_frame()], axis = 1).sort_values('region', ascending = False).copy(deep = True).plot.bar(rot = 0, figsize = (14, 12))
_ = majors_by_region.set(xlabel = 'Regiões', ylabel = 'No. de eventos', ylim = (0, 20)), majors_by_region.legend(['Morte figura importante', 'Episodio com evento', 'Batalhas'], fontsize = 12.)
```
